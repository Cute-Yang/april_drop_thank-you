```python
import pandas as pd
```

```python
file_path = "cwru_mscs.csv"
df = pd.read_csv(file_path, sep=",", encoding="utf-8")
```

```python
df.head(5)
```

```python
group_by_email = df.groupby("email")
# found the index of the path
columns = df.columns
sizeof_columns = len(columns)
column_map = dict(zip(columns, range(sizeof_columns)))
path_feature = "path"
assert path_feature in column_map, "expected column -> {},but not found in our dataframe -> {}".format(path_feature, columns)

# transfer the feature name -> feature indicis
path_indices = column_map[path_feature]
```

```python
drop_duplicated_dfs = []
expected_path_string = "thank-you"
# whether to keep the row which we found contains thank-you for the first time! default is True
keep_thank_you_row: bool = True
for email, sub_df in group_by_email:
    n = len(sub_df)
    drop_duplicated_start = None
    print("email -> {} group has {} datas".format(email, n))
    # found the index while get "thank you" for the first time,and then,drop out the data data after this index!
    for i in range(n):
        path_value = sub_df.iloc[i, path_indices]
        if expected_path_string in path_value:
            # we just keep the drop_indics
            drop_duplicated_start = i
            break
    # select the df we want
    if keep_thank_you_row:
        drop_duplicated_start += 1
    if drop_duplicated_start is not None:
        print("drop duplicated for group which has email -> {}".format(email))
        expected_sub_df = sub_df[:drop_duplicated_start]
        drop_duplicated_dfs.append(expected_sub_df)
    else:
        drop_duplicated_dfs.append(sub_df)

agg_dfs = pd.concat(drop_duplicated_dfs, axis=0)
```

```python
# save the handled data!
save_path = "april.csv"
agg_dfs.to_csv(save_path,sep=",",index=False,encoding="utf-8")
```
